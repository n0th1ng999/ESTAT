# Ler Dataset

```python
dataset_1 = pd.read_csv("../Dados/dataset1_declarations.csv")

dataset_2 = pd.read_csv("../Dados/dataset2_holdings.csv")
```

## Dataset 1

```python
dataset_1.head()
```

```python
dataset_1.info()
```

```python
dataset_1.describe()
```

```python
def summarize_non_numeric_columns(dataset_name, dataset):
    non_numeric_columns = dataset.select_dtypes(exclude="number").columns

    print(f"{dataset_name} - propriedades não numéricas")
    for column in non_numeric_columns:
        unique_values = dataset[column].dropna().unique().tolist()
        preview_values = unique_values[:20]
        suffix = " ..." if len(unique_values) > 20 else ""
        print(f"- {column}: {len(unique_values)} valores únicos")
        print(f"  {preview_values}{suffix}")
    print()


summarize_non_numeric_columns("dataset_1", dataset_1)
```

`agent_id` e `declaration_id` são únicos para cada row no dataset

# Merge Dataset

```python
merged_df = pd.merge(dataset_1, dataset_2,
                     left_index=True, right_index=True,
                     how='inner', suffixes=("","_holdings")
                    )

merged_df.head()
```
